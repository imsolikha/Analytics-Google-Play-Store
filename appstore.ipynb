```python
d1=df[df['type']=='free'].groupby('content_rating')['installs'].mean()
d2=df[df['type']=='paid'].groupby('content_rating')['installs'].mean()
d1.plot(kind='barh')
d2.plot(kind='barh')
```

```python
d=df.pivot_table(index='content_rating',columns='type',values='installs',aggfunc='mean')
```

```python
d.plot(kind='barh',subplots=True)
```

```python
d.plot(kind='barh',subplots=True,layout=(1,2))
```

```python
d.plot(kind='barh',subplots=True,layout=(1,2),sharey=True,grid=True)
```

```python
d.plot(kind='barh',subplots=True,layout=(1,2),sharex=True,grid=True)
```
